```python
import pandas as pd
# data = pd.read_csv('D:/ShipSupplies/DA/data/raw.csv')
data = pd.read_csv('../data/동원산업통합_분류_1_RE.csv', encoding = 'cp949')
```

```python
data.info()
```

```python
df = data.fillna(0)
```

```python
# column정보
df.info()
```

```python
df.isna().count().sum()
```

```python
df1 = df.copy()
# df1
temp1 = df[(df['발주'] == 0) | (df['창고입고'] == 0)] # 발주날짜와 입고 날짜가 없는 데이터
idx = temp1.index
temp1
df1 = df.drop(idx)  # 삭제

df1 = df1.reset_index() # 인덱스 초기화
df1 = df1.drop('index', axis=1) # index 열 삭제

# df1
# df1
df1['발주'] = pd.to_datetime(df1['발주'])
df1['창고입고'] = pd.to_datetime(df1['창고입고'])
```

```python
df1.columns
```

```python
df1['leadtime'] =  df1['창고입고'] - df1['발주'] 
```

```python
# int로 형변환
df1['leadtime'] = df1['leadtime'].apply(lambda x: x.days)
df1['leadtime']
```

```python
t = df1[df1['leadtime'] < 0]
t.index
df1 = df1.drop(t.index)
df1 = df1.reset_index() # 인덱스 초기화
df1 = df1.drop('index', axis=1) # index 열 삭제
df1
```
